```python
import gradio as gr
from groq import Groq

GROQ_API_KEY = "x"

client = Groq(api_key=GROQ_API_KEY)

def chatbot_quimica_fisica(pergunta):
    try:
        prompt_sistema = """
        Você é um especialista em Química e Física.
        Responda de forma clara, objetiva e didática.
        """

        resposta = client.chat.completions.create(
         model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": prompt_sistema},
                {"role": "user", "content": pergunta}
            ],
            temperature=0.5
        )

        return resposta.choices[0].message.content

    except Exception as e:
        return f"Erro encontrado: {str(e)}"

interface = gr.Interface(
    fn=chatbot_quimica_fisica,
    inputs="text",
    outputs="text",
    title="ChatBot Especialista em Química e Física"
)

interface.launch(share=True, debug=True)
```
